```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# Configuration
# ----------------------------
IMAGE_PATH = r'C:/Users/ASUS/Downloads/rose.jpg'  # Replace with your image path

# ----------------------------
# Load Image
# ----------------------------
def load_grayscale_image(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        raise FileNotFoundError(f"Cannot load image from {path}")
    print("âœ… Image loaded successfully.")
    return image

# ----------------------------
# Apply 2D FFT and Compute Spectrum
# ----------------------------
def compute_fft(image):
    f = np.fft.fft2(image)                       # 2D FFT
    fshift = np.fft.fftshift(f)                  # Shift zero frequency to center
    magnitude_spectrum = 20 * np.log(np.abs(fshift) + 1)  # Log scale for visibility
    return magnitude_spectrum

# ----------------------------
# Display Results using Matplotlib
# ----------------------------
def display_results(original, spectrum):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(spectrum, cmap='gray')
    plt.title('Magnitude Spectrum (FFT)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# ----------------------------
# Main Execution
# ----------------------------
def main():
    image = load_grayscale_image(IMAGE_PATH)
    spectrum = compute_fft(image)
    display_results(image, spectrum)

if __name__ == "__main__":
    main()
```
